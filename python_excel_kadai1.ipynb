```python
import openpyxl
from datetime import date

# 新規ワークブックを作成
workbook = openpyxl.Workbook()

# 新規のシートを作成
sheet = workbook.active

# セルに内容を記入する
sheet['B2'] = "請求書"

sheet['B4'] = "株式会社ABC"

sheet['B5'] = "〒101-0022 東京都千代田区神田練塀町300"

sheet['B6'] = "TEL:03-1234-5678 FAX:03-1234-5678"

sheet['B7'] = "担当者名:鈴木一郎 様"

sheet['F4'] = "No."

sheet['F5'] = "日付"

sheet['G4'] = "0001"

# 現在の日付（年月日）を取得して、変数todayに代入する
today = date.today()
date = today.strftime("%Y/%m/%d")
sheet['G5'].value = date

# 表のヘッダー行を書き込む
header = ["商品名", "数量", "単価", "金額"]
sheet['B10'].value = header[0]
sheet['C10'].value = header[1]
sheet['D10'].value = header[2]
sheet['E10'].value = header[3]

# 表の内容を書き込む
sheet['B11'] = "商品A"
sheet['B12'] = "商品B"
sheet['C11'] = 2
sheet['C12'] = 1
sheet['D11'] = 10000
sheet['D12'] = 15000
sheet['E11'] = 20000
sheet['E12'] = 15000
sheet['E13'] = 35000
sheet['B15'] = "小計"
sheet['B16'] = "消費税"
sheet['B17'] = "合計"
sheet['E15'] = 35000
sheet['E16'] = 3500
sheet['E17'] = 38500

# ファイルを保存
workbook.save("請求書_20260612.xlsx")
```
